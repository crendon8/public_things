# Plot daily fills
Demonstration notebook that builds fake daily data and plots sim vs prod fills with pct_diff on a secondary axis.

```python
import pandas as pd
from daily_fills_plot import concat_daily_frames, create_fake_daily_frame, plot_fills_widget

# Define groups in the notebook (7 groups)
FILL_GROUP = {
    "title": "Total fills for {symbol}",
    "primary": (("sim_fills", "sim_fills"), ("prod_fills", "prod_fills")),
    "secondary": (("pct_diff", "pct_diff (%)"),),
}
BUY_GROUP = {
    "title": "Buy fills for {symbol}",
    "primary": (("sim_buy_fills", "sim_buy_fills"), ("prod_buy_fills", "prod_buy_fills")),
    "secondary": (("total_buy_pct_diff", "buy_pct_diff (%)"),),
}
SELL_GROUP = {
    "title": "Sell fills for {symbol}",
    "primary": (("sim_sell_fills", "sim_sell_fills"), ("prod_sell_fills", "prod_sell_fills")),
    "secondary": (("total_sell_pct_diff", "sell_pct_diff (%)"),),
}
BUY_CT_GROUP = {
    "title": "Buy counts for {symbol}",
    "primary": (("sim_buys", "sim_buys"), ("prod_buys", "prod_buys")),
    "secondary": (("ct_buy_pct", "ct_buy_pct (%)"),),
}
SELL_CT_GROUP = {
    "title": "Sell counts for {symbol}",
    "primary": (("sim_sells", "sim_sells"), ("prod_sells", "prod_sells")),
    "secondary": (("ct_sell_ct", "ct_sell_ct (%)"),),
}
BUY_PX_GROUP = {
    "title": "Buy prices for {symbol}",
    "primary": (("sim_buy_px", "sim_buy_px"), ("prod_buy_px", "prod_buy_px")),
    "secondary": (("buy_bps_diff", "buy_bps_diff (bps)"),),
}
SELL_PX_GROUP = {
    "title": "Sell prices for {symbol}",
    "primary": (("sim_sell_px", "sim_sell_px"), ("prod_sell_px", "prod_sell_px")),
    "secondary": (("sell_bps_diff", "sell_bps_diff (bps)"),),
}
```

```python
dates = pd.date_range("2024-01-01", periods=14, freq="D")

# Build one dataframe per day (no date column yet)
daily_frames = [create_fake_daily_frame(date=d, names=["ABC", "DEF", "XYZ"]) for d in dates]

# Attach the date during concat so downstream plots have it
df = concat_daily_frames(daily_frames, dates=dates)

# Inspect the first few rows
df.head()
```

```python
widget = plot_fills_widget(df, groups=[
    ("fills", FILL_GROUP),
    ("buy", BUY_GROUP),
    ("sell", SELL_GROUP),
    ("buy_ct", BUY_CT_GROUP),
    ("sell_ct", SELL_CT_GROUP),
    ("buy_px", BUY_PX_GROUP),
    ("sell_px", SELL_PX_GROUP),
])
widget
```

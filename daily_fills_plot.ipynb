# Plot daily fills
Demonstration notebook that builds fake daily data and plots sim vs prod fills with pct_diff on a secondary axis.

```python
import pandas as pd
from daily_fills_plot import concat_daily_frames, create_fake_daily_frame, plot_fills
```

```python
dates = pd.date_range("2024-01-01", periods=14, freq="D")

# Build one dataframe per day (no date column yet)
daily_frames = [create_fake_daily_frame(date=d, names=["ABC", "DEF", "XYZ"]) for d in dates]

# Attach the date during concat so downstream plots have it
df = concat_daily_frames(daily_frames, dates=dates)

# Inspect the first few rows
df.head()
```

```python
fig = plot_fills(df, name="ABC")
fig
```

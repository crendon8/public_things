# attach_buy_sell demo

Quick example showing how to pull buy/sell prices from the market frame onto the fills frame using the helper added to `mini_symbol_plot.py`.

```python
import pandas as pd

from mini_fake_data import make_fake_frames
from mini_symbol_plot import attach_buy_sell

fills_df, market_df = make_fake_frames()
fills_df.head()
```

```python
# Attach INTERNAL fair-value bid/ask columns to the fills.
# Column naming pattern: "{prefix}{symbol}{buy_suffix}" / "{prefix}{symbol}{sell_suffix}".
augmented = attach_buy_sell(
    fills_df,
    market_df,
    prefix="INTERNAL.",
    buy_suffix="_FV_bid",
    sell_suffix="_FV_ask",
)
augmented[["symbol", "rxTime", "buy", "sell", "price", "size"]].head()
```
